```python
#SINGLE INHERITENCE

class mother:
    def fun1(self,name,age):
        self.name=name
        self.age=age

class father(mother):
    def fun2(self,fname,fage):
        self.fname=fname
        self.fage=fage
        print(self.name)
        print(self.age)
        print(self.fname)
        print(self.fage)
k1=father()
k1.fun1('chandra',45)
k1.fun2('jayavel',48)
```

```python
#MULTIPLE INHERITANCE

class family:
    def fun1(self,persons):
        self.persons=persons

class grandpa:
    def fun2(self,gname):
        self.gname=gname

class grandma:
    def fun3(self,gpname):
        self.gpname=gpname

class father:
    def fun4(self,fname):
        self.fname=fname

class mother:
    def fun5(self,mname):
        self.mname=mname

class sister:
    def fun6(self,sname):
        self.sname=sname

class brother:
    def fun7(self,bname):
        self.bname=bname

class brother1:
    def fun8(self,b1name):
        self.b1name=b1name

class me(family,grandpa,grandma,father,mother,sister,brother,brother1):
    def son(self):
        print('total family members:',self.persons)
        print('grandpa name:',self.gname)
        print('grandma name:',self.gpname)
        print('father name:',self.fname)
        print('mother name:',self.mname)
        print('sister name:',self.sname)
        print('brother name:',self.bname)
        print('brother1 name:',self.b1name)
        
k=me()
k.fun1(8)
k.fun2('seerangan')
k.fun3('kamalam')
k.fun4('jayavel')
k.fun5('chandra')
k.fun6('vaideki')
k.fun7('vinodh')
k.fun8('aravid')
k.son()
```

```python
#MULTILEVEL INHERITANCE

class family:
    def fun1(self,persons):
        self.persons=persons

class grandpa(family):
    def fun2(self,gname):
        self.gname=gname

class grandma(grandpa):
    def fun3(self,gpname):
        self.gpname=gpname

class father(grandma):
    def fun4(self,fname):
        self.fname=fname

class mother(father):
    def fun5(self,mname):
        self.mname=mname

class sister(mother):
    def fun6(self,sname):
        self.sname=sname

class brother(sister):
    def fun7(self,bname):
        self.bname=bname

class brother1(brother):
    def fun8(self,b1name):
        self.b1name=b1name

class me(brother1):
    def son(self):
        print('total family members:',self.persons)
        print('grandpa name:',self.gname)
        print('grandma name:',self.gpname)
        print('father name:',self.fname)
        print('mother name:',self.mname)
        print('sister name:',self.sname)
        print('brother name:',self.bname)
        print('brother1 name:',self.b1name)
        
k=me()
k.fun1(8)
k.fun2('seerangan')
k.fun3('kamalam')
k.fun4('jayavel')
k.fun5('chandra')
k.fun6('vaideki')
k.fun7('vinodh')
k.fun8('aravid')
k.son()
```
